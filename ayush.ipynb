```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pytz
import folium
from geopy.geocoders import Nominatim
from datetime import datetime
```

```python
data = pd.read_csv('EVChargingStationUsage.csv')
```

```python
print(data.columns)
```

```python
# Select the target variable and features
X = data.drop(columns=['Energy (kWh)'])  # Features
y = data['Energy (kWh)']  # Target

```

```python
# Create a correlation matrix heatmap
corr_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Visualize specific relationships with scatter plots
plt.figure(figsize=(15, 5))

# Example 1: Energy vs. GHG Savings
plt.subplot(1, 2, 1)
sns.scatterplot(data=data, x='Energy (kWh)', y='GHG Savings (kg)')
plt.xlabel('Energy (kWh)')
plt.ylabel('GHG Savings (kg)')
plt.title('Energy vs. GHG Savings')
```

```python
# Example 2: Energy vs. Latitude
plt.subplot(1, 2, 2)
sns.scatterplot(data=data, x='Energy (kWh)', y='Latitude')
plt.xlabel('Energy (kWh)')
plt.ylabel('Latitude')
plt.title('Energy vs. Latitude')
```
